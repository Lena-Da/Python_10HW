```python
import math
from sympy import *
```

```python
x = Symbol('x')
y = (sin(x)/2 + cos(x)/2)**2 * sin(x)
y
```

```python
plot(y)
```

```python
roo=solve(y)
print('Корни ', roo)
```

```python
dif=diff(y)
print ('Дифференциал ', dif)
res=solve(dif)
print('Корни дифференциала ', res)
```

```python
plot(dif)
```

```python
roo=solve(y)
interval_y = interv(roo)
for i in interval_y:
    start = -oo if interval_y.index(i) == 0 else i[0]
    end = oo if interval_y.index(i) == len(interval_y)-1 else i[1]
    if y.evalf(subs={x:(i[0] + i[1])/2}) > 0:
        print(f'f > 0 в промежутке {start, end}')
    else:
        print(f'f < 0 в промежутке {start, end}')
```
